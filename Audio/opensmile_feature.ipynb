```python
import os
import time

import numpy as np
import pandas as pd

# import audb
import audiofile
import opensmile
```

```python
# declarion of opensmile configuration

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
```

```python
# Path where all audio files are stored
path = "/home/pulkit_2111mt05/Flickscore_final/Dataset/audio"
audio_files = os.listdir(path)
```

```python
audio_emb_dict = {}
count = 1
for f in audio_files:
    print("count: ",count)
    file = os.path.join(path,f)
#     print(file)
    audio_features = []


    signal, sampling_rate = audiofile.read(
    file,
    always_2d=True,
    )
    
    features = smile.process_signal(
    signal,
    sampling_rate
    )

    audio_features = [features[c][0]for c in features.columns]
    audio_emb_dict[f] = audio_features
    count = count + 1
```

```python
audio_emb_df = pd.DataFrame.from_dict(audio_emb_dict)
audio_emb_df = audio_emb_df.T
```

```python
audio_emb_df
```

```python
for column in audio_emb_df.columns:
    srt = audio_emb_df[column].copy()
    srt = list(set(srt))
    srt.sort()
    srt = np.array(srt)
    srt = [i for i in srt if i not in [-np.inf, np.inf]]

    audio_emb_df[column] = audio_emb_df[column].replace([-np.inf], min(srt))       
    audio_emb_df[column] = audio_emb_df[column].replace([np.inf], max(srt))  
    audio_emb_df[column] = audio_emb_df[column].fillna(min(srt))

audio_emb_df_min_max = audio_emb_df.copy()
i = 0
# apply normalization techniques
for column in audio_emb_df_min_max.columns:

    audio_emb_df_min_max[column] = (audio_emb_df[column] - audio_emb_df[column].min()) / (audio_emb_df[column].max() - audio_emb_df[column].min())
    audio_emb_df_min_max[column] = audio_emb_df_min_max[column].fillna(0)
    
```
