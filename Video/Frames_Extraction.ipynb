```python
import numpy as np
from PIL import Image
import os
import cv2
from sklearn.cluster import KMeans

import os
#import helpers
import imageio

import numpy as np
from skimage.feature import greycomatrix

from tqdm import tqdm

import shutil
```

## Frames Extraction

```python
def get_frames(path):
    parent_dir = os.listdir(path)
    parent_save = "/home/pulkit_2111mt05/save"
    movie_names = []

    for child_dir in parent_dir :
        os.mkdir(os.path.join(parent_save,child_dir))       
        count = 1     
        frame_path = parent_save+"/"+child_dir
        #os.mkdir(frame_path)

        video_path = os.path.join("/home/pulkit_2111mt05/trailer_more_5000" ,child_dir)
        movie_names.append(child_dir)    
        video_reader = cv2.VideoCapture(video_path)
        #frames_req = 5000           #No. of frame requires from each video 
        video_frames_count = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))  #COUNTING NO. OF FRAMES IN A VIDEO
        print(child_dir," ",video_frames_count)


        skip_frames = 1

        for frame_id in range(video_frames_count):
            video_reader.set(cv2.CAP_PROP_POS_FRAMES, frame_id * skip_frames)
            ret, frame = video_reader.read()
            save_path = os.path.join(frame_path,"frame"+ str(frame_id)+".jpg")        
            cv2.imwrite(save_path, frame)

        count += 1        
        print(count) 
        
```

## K_Means

```python

def get_K_Frames(path):
    
    IMG_SIZE = 480
    count = 28
    for movie_name in os.listdir(path)[count:]:
        data = []
        movie_namex , idx  = movie_name.split("_")
        idx = idx.split(".")[0]
        print(count, movie_name)  
        for frames in os.listdir(path+movie_name):        
            img=cv2.imread(path+movie_name+"/"+frames)    
            img=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
            img=img.astype('float32')

            data.append(img)
        data = np.array(data)

        data = data/255.0
        reshaped_data = data.reshape(len(data),-1)

        kmeans = KMeans(n_clusters=5, random_state=0)
        clusters = kmeans.fit_predict(reshaped_data)
        os.mkdir("/home/pulkit_2111mt05/FickScore/k_frames2/"+idx)
        for i in range(0,5):
            nm = kmeans.cluster_centers_[i].reshape(480,480,3)
            im = Image.fromarray((nm * 255).astype(np.uint8))
            #im = Image.fromarray(nm)
            im.save("/home/pulkit_2111mt05/FickScore/k_frames2/"+idx+"/"+idx+"_"+movie_namex+ str(i) + ".jpg")

        count +=1
  
```

```python
def get_entropy(image):

    img = imageio.imread(image)
    img_r= img[:, :, 0]
    img_g= img[:, :, 1]
    img_b= img[:, :, 2]
    img_gray = 0.2989 * img_r + 0.5870 * img_g + 0.1140 * img_b
    img_gray= img_gray.astype(np.uint8)
    


    glcm = np.squeeze(greycomatrix(img_gray, distances=[1], 
                               angles=[0], symmetric=True, 
                               normed=True))
    print(glcm.shape)
    entropy = -np.sum(glcm*np.log2(glcm + (glcm==0)))

    return entropy
```
